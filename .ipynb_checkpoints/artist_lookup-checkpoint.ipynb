```python
import pickle
import pandas as pd
```

```python
features_from_spotify = pickle.load(open('features_from_spotify.pkl'))
```

```python
features_from_training_data = pickle.load(open('features_from_training_data.pkl'))
```

```python
assert(len(features_from_spotify) == len(features_from_training_data))
```

```python
artist_lookup = features_from_training_data.copy()
for key in artist_lookup:
    artist_lookup[key].update(features_from_spotify[key])
```

```python
pd.DataFrame(artist_lookup)
```

```python
pickle.dump(artist_lookup, open('artist_lookup.pkl', 'w'))
```
