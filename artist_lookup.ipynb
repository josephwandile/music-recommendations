```python
import pickle
import pandas as pd
```

```python
spotify = pickle.load(open('spotify.pkl'))
manual = pickle.load(open('artistic_features.pkl'))
```

```python
assert(len(spotify) == len(manual))
```

```python
artist_lookup = manual.copy()
for key in artist_lookup:
    artist_lookup[key].update(spotify[key])
```

```python
pd.DataFrame(artist_lookup)
```

```python
pickle.dump(artist_lookup, open('artist_lookup.pkl', 'w'))
```
